## Stack

```python
sample = ')(())'
stack = []
VPS = True
for i in sample:
  if i == '(':
    stack.append(i) # 
  else:
    if stack:
      stack.pop()
    else:
      VPS = False
print(VPS)
```

```python
for _ in range(int(input())):
  stack = []
  VPS = True
  for i in input():
    if i == '(':
      stack.append(i) # 
    else:
      if stack:
        stack.pop()
      else:
        VPS = False
        break
  if stack:
    VPS = False
  print('YES' if VPS else 'NO')
```

## Que

```python
from collections import deque

N = int(input())
q = deque(range(1, N + 1))

while len(q) > 1:
  q.popleft()
  q.append(q.popleft())
print(q)
```

## map

```python
d = {}

for _ in range(int(input())):
  book = input()
  if book in d:
    d[book] += 1
    # d[book] = d[book] + 1
  else:
    d[book] = 1

max_value = max(d.values())
max_title_list = []
for k, v in d.items():
  if v == max_value:
    max_title_list.append(k)
print(sorted(max_title_list)[0])
```
