## DFS

```python
N, M = map(int, input().split())
adj = [[0] * N for _ in range(N)] # N x N

for _ in range(M):
  a, b = map(lambda x: x - 1, map(int, input().split()))
  adj[a][b] = adj[b][a] = 1

visited = [False] * N
def dfs(now):
  for next in range(N):
    if adj[now][next] and not visited[next]:
      visited[next] = True
      dfs(next)
count = 0
for i in range(N):
  if not visited[i]:
    count += 1
    visited[i] = True
    dfs(i)
print(count)
```

```python
from collections import deque

N, M = map(int, input().split())
board = [input() for _ in range(N)]
# (1, 0), (-1, 0), (0, -1), (0, 1)
dx = (1,0,-1,0) # 우, 상, 좌, 하
dy = (0,1,0,-1) # 우, 상, 좌, 하


def is_valid_coord(y, x): # board size < N X M
  return (0 <= y < N) and (0 <= x < M)

def bfs():
  chk = [[False] * M for _ in range(N)]
  chk[0][0] = True
  dq = deque()
  dq.append((0,0,1))
  while dq:
    y, x, d = dq.popleft() # y, x, 이동횟수
    if y == N-1 and x == M-1: # 목적지에 도착했을 때
      return d
    nd = d + 1
    # Move
    for k in range(4): # 상하좌우
      ny = y + dy[k]
      nx = x + dx[k]
      if is_valid_coord(ny, nx) and board[ny][nx] == '1' and not chk[ny][nx]:
        # print(y, x, d)
        chk[ny][nx] = True
        dq.append((ny, nx, nd))
print(bfs())
```

## 이진탐색

```python
# quiz (1)
N = int(input()) # 지방의 수
req = list(map(int, input().split())) # 지방에서 요청한 금액
M = int(input()) # 국가 총 예산

low = 0
high = max(req)
mid = (low + high) // 2 # 상한액(target)

while low <= high:
  req_sum = sum(min(mid, r) for r in req)
  print(f'low : {low}, mid : {mid},high : {high}')
  print(f'target : {req_sum}, total : {M}')
  if req_sum <= M:
    low = mid + 1
  else:
    high = mid - 1
  mid = (low + high) // 2
print(mid)
```

```python
# quiz (2)
N = int(input())
N_list = sorted(list(map(int, input().split())))

M = int(input())
M_list = list(map(int, input().split()))

for target in M_list:
  find = False
  low = min(N_list) # 0 
  high = max(N_list)
  while low <= high:
    mid = int((low + high) / 2)
    if mid > target:
      high = mid - 1
    elif mid < target:
      low = mid + 1
    else:
      find = True 
      break
  if find:
    print(1)
  else:
    print(0)
```
