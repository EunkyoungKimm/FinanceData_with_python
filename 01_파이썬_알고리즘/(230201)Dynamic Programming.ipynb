## 피보나치

```python
count = 0

def f(n):
  # global count
  # count += 1
  if n == 0 :
    return 0
  elif n == 1:
    return 1
  
  return f(n-1) + f(n-2)
f(int(input()))
# print(count)
```

```python
# Top - Down(재귀)
cache = [-1] * 91
cache[0] = 0
cache[1] = 1
# count = 0

def f(n):
  # global count
  # count += 1
  if cache[n] == -1:
    cache[n] = f(n-1) + f(n-2)
  return cache[n]
f(int(input()))
# print(count)
```

```python
# Bottom - Up (반복문)
N = int(input())
cache = [-1] * 91
cache[0] = 0
cache[1] = 1
# count = 0

for i in range(2, N + 1):
  # global count
  # count += 1
  cache[i] = cache[i - 1] + cache[i - 2]
print(cache[N])
# print(count)
```

## Combination

```python
# Top - Down(재귀)
N, K = map(int, input().split())
cache = [[-1] * 1001 for _ in range(1001)]

def f(n, k):
  if cache[n][k] != -1:
    return cache[n][k] 
  
  if k == 0 or k == n:
    cache[n][k] = 1
  else:
    cache[n][k] = f(n-1, k-1) + f(n-1, k)
    cache[n][k] %= 10007
  return cache[n][k]
print(f(N, K))
```

```python
# Bottom - Up (반복문)
N, K = map(int, input().split())
cache = [[-1] * 1001 for _ in range(1001)]

for n in range(1001):
  cache[n][0] = cache[n][n] = 1
  for k in range(1,n): # ex. 1,5 -> [1,2,3,4]
    cache[n][k] = cache[n-1][k-1] + cache[n-1][k]
    cache[n][k] %= 10007
cache[N][K] 
```

```python
# Quiz (2) 타일문제
n = int(input())
cache = [-1] * 1001
cache[1] = 1
cache[2] = 2
for i in range(3, 1001):
  cache[i] = (cache[i-1] + cache[i-2]) % 10007
print(cache[n])
```

```python
# Quiz(3) 계단문제
cache = [[0] * 10 for _ in range(101)] # n X N

for d in range(1, 10): # 길이가 1이고 마지막숫자가 1~9
  cache[1][d] = 1

for n in range(2, 101):
  for d in range(10):
    if d == 0: # 0이 올때 cache[n-1][d+1]
      cache[n][d] = cache[n-1][d+1]
    elif d == 9: # 9가올 때 cache[n-1][d-1]
      cache[n][d] = cache[n-1][d-1]
    else:
      cache[n][d] = cache[n-1][d-1] + cache[n-1][d+1]
    cache[n][d] %= 1_000_000_000

N = int(input())
result = 0
for d in range(10):
  result += cache[N][d]
  result %= 1_000_000_000
print(result)
```
