## apply
- 굉장히 중요하다!! --> 잘 사용할 수 있으면 코딩 시간이 확 준다!

```python
df_kospi = fdir.StockListing('KOSPI')
df_kospi.head()
```

```python
# 오른주식(변화율 양수), 내린주식(변화율 음수)을 새로운 칼럼으로 선언
# 방법 3가지 > 람다 쓰도록 연습하자
# lambda로 함수를 만들어서 활용
df_kospi['방법1'] = df_kospi['ChagesRatio'].apply(lambda x: '오른주식' if x > 0 else '내린주식') # apply: 내가 정의 하는 칼럼(ChagesRatio)에 함수가 적용됨
df_kospi.head()
```

```python
def check_chagesratio(data):
    if data > 0:
        return '오른주식'
    else:
        return '내린주식'
```

```python
df_kospi['방법2'] = df_kospi['ChagesRatio'].apply(check_chagesratio)
df_kospi.head()
```

```python
import numpy as np
df_kospi['방법3'] = np.where(df_kospi['ChagesRatio'] > 0, '오른주식', '내린주식')
df_kospi.head()
```

```python
# value_counts() --> 많이 쓴다!
(df_kospi['방법1'] == df_kospi['방법2']).value_counts()
```

```python
(df_kospi['방법2'] == df_kospi['방법3']).value_counts()
```

```python
df_kospi['방법1'].value_counts()
```

```python
data = [
    [1000, 1100, 900, 1200, 1300],
    [800, 2000, 1700, 1500, 1800]
]
index = ['자본금', '부채']
columns = ["2020/03", "2020/06", "2020/09", "2021/03", "2021/06"]
df = pd.DataFrame(data, index, columns)
df
```

```python
# 방법1 - apple, lambda 사용
result = df.stack().reset_index().rename(columns = {'level_0' : '계정', 'level_1' : '년월', 0 : '금액'})
result['연도'] = result['년월'].apply(lambda x: x.split('/')[0])
result['월'] = result['년월'].apply(lambda x: x.split('/')[1])
result
```

```python
# 방법2 - concat
result = df.stack().reset_index().rename(columns = {'level_0' : '계정', 'level_1' : '년월', 0 : '금액'})
date = pd.DataFrame(list(result['년월'].str.split('/')), columns = ['연도', '월']) # 여긴 str해야되고 위에는 안해야됨?!
result = pd.concat([result, date], axis = 1)
result
```

```python
# 헷갈리는 점
result['년월'].str # 시리즈에서 str하면 str에서 쓰는 함수 사용 가능(통째로 적용) / apply는 '하나하나, 한줄한줄'를 별개의 것으로 하나하나 적용!
```

```python
# unstack 좀 더 알아보기
result_group = result.groupby(['계정', '연도']).sum()
result_group
```

## Pivot & Pivot_table
- groupby랑 비슷한 느낌
- 어떤 점이 다른지 구분하기!

```python
# 데이터를 빠르게 볼때는 groupby / 데이터를 재정렬하고 싶을 때는 pivot_table 사용하면 good(원래 인덱스..밸류를 다른거 넣고 싶을때 사용하면 좋게 사용하는거라고 하심)
# 보통은 groupby / index, columns, values에 뭐가 들어갈지 정확하게 알때는 pivot_table
# groupby 사용하는게 무난하긴 함!
```

```python
data = [
    ["2022-08-12", "A전자", 77000],
    ["2022-08-13", "A전자", 74400],
    ["2022-08-12", "B전자", 153000],
    ["2022-08-13", "B전자", 150500],
    ["2022-08-12", "C반도체", 100500],
    ["2022-08-13", "C반도체", 101500]
]
columns = ["날짜", "종목명", "종가"]
df = pd.DataFrame(data=data, columns=columns)
df
```

```python
pd.pivot_table(index = '날짜', columns = '종목명', values = '종가', data = df)
```

```python
pd.pivot_table(index = '날짜', values = '종가', data = df, aggfunc = 'mean')
```
