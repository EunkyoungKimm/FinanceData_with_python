# Matplotlib
- figure, ax 를 다음과 같이 이해하면 좋다.
    - figure = 도화지
    - ax = 도화지 안에 x축,y축이 ax(그래프..)(figure안에 axes를 올린다)

## 시각화

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.plot()
```

```python
plt.figure(figsize = (10, 5))
plt.plot()
```

```python
# numpy를 활용해 임의의 데이터 만들기
import numpy as np
x = np.arange(100,300,10)
y = np.random.rand(20)
plt.plot(x, y)
plt.xlabel('x-Axis')
plt.ylabel('y-Axis')
plt.title('title')
```

```python
plt.figure(figsize = (10,5))
plt.plot(x, y, color='r', linestyle='--', marker='o') # plt.plot(x, y, 'ro--'')
```

## 그래프 여러개 그리는 방법(1)

```python
fig = plt.figure(figsize = (16, 10))

ax1 = fig.add_subplot(2,2,1) # 2x2 사이즈에서 첫번째
ax1.plot(x, y, color='r', linestyle='--')

ax2 = fig.add_subplot(2,2,2) # 2x2 사이즈에서 두번째
ax2.hist(y)

ax3 = fig.add_subplot(2,2,3) 
ax3.scatter(x, y)
```

## 그래프 여러개 그리는 방법(2)

```python
fig, axes = plt.subplots(2, 2, figsize = (16, 10))
# print(axes.shape) # (2, 2)

axes[0,0].plot(y)
axes[1,0].hist(y)
axes[0,1].scatter(x, y)
axes[1,1].plot(y, 'g--')
```
