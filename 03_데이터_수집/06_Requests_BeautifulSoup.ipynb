```python
name_list = []
title_list = []
price_list = []
opinion_list = []
source_list = []
date_list = []

for i in range(30): 
    price_idx = 4 * i
    opinion_idx = 4 * i + 1
    source_idx = 4 * i + 2
    date_idx = 4 * i + 3
    
    name = table.find_all('a')[2*i].text
    title = table.find_all('p')[i].text.strip() # a태그 > 2 * i + 1 해도 됨
    price = table.find_all('div', attrs = {'class': 'line3'})[price_idx].text.strip()[4:].strip()
    opinion = table.find_all('div', attrs = {'class': 'line3'})[opinion_idx].text.strip()
    source = table.find_all('div', attrs = {'class': 'line3'})[source_idx].text.strip()
    date = table.find_all('div', attrs = {'class': 'line3'})[date_idx].text.strip()
    
    
    name_list.append(name)
    title_list.append(title)
    price_list.append(price)
    opinion_list.append(opinion)
    source_list.append(source)
    date_list.append(date)
    
print(name_list)
print(title_list)
print(price_list)
print(opinion_list)
print(source_list)
print(date_list)
```

```python
import pandas as pd
df = pd.DataFrame({'종목' : name_list, '제목' : title_list, '적정가격' : price_list, '투자의견' : opinion_list, '제공출처' : source_list, '작성일' : date_list})
df
```
