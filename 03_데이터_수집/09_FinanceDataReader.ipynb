## API와 라이브러리 다른점?
API(Application Programming Interface)는 소프트웨어와 애플리케이션을 빌드하기 위한 프로토콜, 루틴 및 도구의 집합입니다. 소프트웨어 구성 요소가 상호 작용하는 방법을 지정하고 API는 서로 다른 시스템과 플랫폼 간의 통신을 허용합니다.

반면에 라이브러리는 개발자들이 API에 연결하거나 데이터를 구문 분석하는 것과 같은 일반적인 기능을 수행하기 위해 자신의 응용 프로그램에서 사용할 수 있는 미리 작성된 코드의 모음이다. 라이브러리는 특정 프로그래밍 언어나 플랫폼을 위해 더 많은 추상화와 편리한 기능을 제공하는 더 높은 수준의 API라고 생각할 수 있다.

간단히 말해서, API는 두 시스템이 상호 작용하는 방식을 정의하는 반면, 라이브러리는 공통적인 문제에 대한 준비된 솔루션을 제공한다.
