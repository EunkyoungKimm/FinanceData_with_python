## 배당정보를 수집하고 가공

- 배당성향: 50 > 100만원 벌면 50%는 주인에게 감
- 용어나 경제 개념들은 더 찾아보기!

```python
stock = dart.report('033780', '배당', 2021, '11011') # KT&G > 케이티앤지
배당성향 = stock[stock['se'] == '(연결)현금배당성향(%)']['thstrm'].iloc[0] # 현금배당수입률(%) 주당 현금배당금(원)
배당수익률 = stock[stock['se'] == '현금배당수익률(%)']['thstrm'].iloc[0]
배당금 = stock[stock['se'] == '주당 현금배당금(원)']['thstrm'].iloc[0]
print(배당성향, 배당수익률, 배당금)
```

```python
def get_div(cor, year):
    stock = dart.report(cor, '배당', year, '11011') 
    배당성향 = stock[stock['se'] == '(연결)현금배당성향(%)']['thstrm'].iloc[0] # 현금배당수입률(%) 주당 현금배당금(원)
    배당수익률 = stock[stock['se'] == '현금배당수익률(%)']['thstrm'].iloc[0]
    배당금 = stock[stock['se'] == '주당 현금배당금(원)']['thstrm'].iloc[0]
    return 배당성향, 배당수익률, 배당금
```

### ROE

```python
result_int['평균자기자본'] = result_int['자본총계'].rolling(2).mean() # 이동평균 활용(rolling)
result_int['ROE'] = result_int['당기순이익'] / result_int['평균자기자본'] * 100
result_int
```

```python
def get_finance_data(stork_name, start_year, end_year):
  dates = list(range(start_year, end_year + 1))
  result = pd.DataFrame()
  for year in dates:
    data = dart.finstate(stork_name, year, '11011')
    account_nm_list = ['자산총계', '부채총계', '자본총계', '매출액', '영업이익', '당기순이익']
    data = data[data['account_nm'].isin(account_nm_list)]
    subset_data = data[['stock_code','bsns_year' ,'account_nm', 'thstrm_amount','fs_nm']]
    pivot = pd.pivot_table(data = subset_data, index = 'bsns_year', columns = 'account_nm', values = 'thstrm_amount', aggfunc='first')
    result = result.append(pivot)

  # 원화 , 처리
  result_int = result.apply(lambda x: x.str.replace(',','')).astype(np.int64)
  result_int['매출증가율'] = result_int['매출액'].diff() / result_int['매출액'] * 100
  result_int['부채비율'] = result_int['자본총계'] / result_int['자산총계'] * 100
  result_int['당기순이익증가율'] = (result_int['당기순이익'].diff() / result_int['당기순이익']) * 100
  result_int['ROA'] = (result_int['당기순이익'] / result_int['자산총계']) * 100
  result_int['영업이익증가율'] = (result_int['영업이익'].diff() / result_int['영업이익']) * 100
  result_int['평균자기자본'] = result_int['자본총계'].rolling(2).mean() # 이동평균 활용
  result_int['ROE'] = result_int['당기순이익'] / result_int['평균자기자본'] * 100
  return result_int.drop(['평균자기자본'], axis = 1) # 평균자기자본은 계산을 위해서만 필요
```
