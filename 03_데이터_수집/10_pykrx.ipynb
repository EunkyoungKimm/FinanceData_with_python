```python
from pykrx import stock
```

```python
# 티커 리스트
tickers = stock.get_market_ticker_list('20230207')
tickers
```

```python
tickers = stock.get_market_ticker_list(market = 'KOSDAQ') # 코스닥에 있는 종목들만
print(tickers)
```

```python
t1 = stock.get_market_ticker_list('20030227')
t2  = stock.get_market_ticker_list('20230227')
print(len(t1), len(t2)) # 종목 수
```

```python
intersection = set(t1) & set(t2)
print(len(intersection)) # 현재까지 살아남은 기업 > 아무 기업에 투자하면 x (20년 후에 살아남을지..)
```

```python
diff = set(t2) - set(t1) # 2003년 이후 상장
print(len(diff))
print(diff)
```

```python
diff = set(t1) - set(t2) # 상장폐지 > 꽤많음.. 재무재표 보는거 중요해! 분석해야 해!
print(len(diff))
print(diff)
```

```python
market_fundamental = stock.get_market_fundamental('20230206') # 이렇게 fundamental도 가져올 수 있음!(financedatareader에서는 없었던..)
market_fundamental.head()
```

```python
market_fundamental.sort_values('PER') # per: eps에 비해서.. 돈 버는거에 비해서 주가가 얼마나 높은지 >> 주가가 높다(주가가격 비싸다>비싸게 팔린다>왜? 인기 많으니까) or 미래에 발전 가능성 있다..
```

```python
# per들의 분포들을 잘 보여주는 그래프 > 히스토그램
# plotly histogram 구글 쳐서 적절해 보이는 그래프의 소스코드 가져오면 됨

import plotly.express as px
df = px.data.tips()
fig = px.histogram(market_fundamental, x="PER")
fig.show()
```

```python
market_fundamental[market_fundamental['PER'] < 75].sort_values('PER')
```
